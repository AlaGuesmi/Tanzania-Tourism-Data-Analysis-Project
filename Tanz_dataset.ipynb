```python
#identifying columns that contain missing values
mv=df.isnull().sum()
mv
```

```python
df['travel_with']=df['travel_with'].fillna(df['travel_with'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['most_impressing']=df['most_impressing'].fillna(df['most_impressing'].value_counts().idxmax())
```

```python
df.isnull().sum()
```

```python
for col in ['total_female','total_male']:
    df[col]=df[col].fillna(df[col].mean())
```
